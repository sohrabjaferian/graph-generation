## Some Evaluations

```python
from eval.stats import degree_stats, clustering_stats, orbit_stats_all
import networkx as nx
import utils

# Load real and synthetic graphs
real_graphs = utils.load_graph_list('./real_data/train_data.dat', is_real=True)
synthetic_graphs = utils.load_graph_list('./synthetic_data/GraphRNN_RNN_KEGG_4_128_pred_100_1.dat', is_real=False)

# Trim to same length
synthetic_graphs = synthetic_graphs[:len(real_graphs)]

# Clean duplicate edges and self-loops
def clean_graphs(graphs):
    cleaned = []
    for G in graphs:
        G = nx.Graph(G)  # make simple undirected (deduplicates edges)
        G.remove_edges_from(nx.selfloop_edges(G))
        cleaned.append(G)
    return cleaned

real_graphs = clean_graphs(real_graphs)
synthetic_graphs = clean_graphs(synthetic_graphs)

# Eval
mmd_degree = degree_stats(real_graphs, synthetic_graphs)
mmd_clustering = clustering_stats(real_graphs, synthetic_graphs)
try:
    mmd_orbits = orbit_stats_all(real_graphs, synthetic_graphs)
except Exception as e:
    print("ORCA failed:", e)
    mmd_orbits = -1

print("-------------------------")
print("MMD Degree:", mmd_degree)
print("MMD Clustering:", mmd_clustering)
print("MMD 4-Orbits:", mmd_orbits)
```

```python
import random
random.seed(0)

real = load_graph_list('./real_data/train_data.dat', is_real=True)
gen  = load_graph_list('./synthetic_data/GraphRNN_RNN_KEGG_4_128_pred_100_1.dat', is_real=False)

# uniform subsample synthetic -> same count as real
gen = random.sample(gen, k=len(real))

m = mean_graph_discrepancy(real, gen)
for k, v in m.items():
    print(f"{k}: {v:.4f}")
```

## Plots

```python
import matplotlib.pyplot as plt
import networkx as nx
from utils import load_graph_list

real = load_graph_list('./real_data/train_data.dat', is_real=True)
gen  = load_graph_list('./synthetic_data/GraphRNN_RNN_KEGG_4_128_pred_100_1.dat', is_real=False)

# pick first few examples
n_show = 3
plt.figure(figsize=(12, 6))

for i in range(n_show):
    plt.subplot(2, n_show, i + 1)
    nx.draw_spring(real[i], node_size=30, alpha=0.8, with_labels=False)
    plt.title(f"Real #{i}")

for i in range(n_show):
    plt.subplot(2, n_show, n_show + i + 1)
    nx.draw_spring(gen[i], node_size=30, alpha=0.8, with_labels=False)
    plt.title(f"Generated #{i}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def degree_list(graphs):
    return [d for G in graphs for _, d in G.degree()]

def clustering_list(graphs):
    return [v for G in graphs for v in nx.clustering(G).values()]

deg_real = degree_list(real)
deg_gen  = degree_list(gen)
clust_real = clustering_list(real)
clust_gen  = clustering_list(gen)

plt.figure(figsize=(12,5))

# Degree distribution
plt.subplot(1,2,1)
plt.hist(deg_real, bins=20, alpha=0.6, label='Real')
plt.hist(deg_gen, bins=20, alpha=0.6, label='Generated')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.legend()
plt.title('Degree Distribution')

# Clustering coefficient distribution
plt.subplot(1,2,2)
plt.hist(clust_real, bins=20, alpha=0.6, label='Real')
plt.hist(clust_gen, bins=20, alpha=0.6, label='Generated')
plt.xlabel('Clustering Coefficient')
plt.ylabel('Frequency')
plt.legend()
plt.title('Clustering Distribution')

plt.tight_layout()
plt.show()
```
