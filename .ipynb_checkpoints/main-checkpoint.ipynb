```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import itertools
import random
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib
from torch_geometric.nn import GCNConv
from torch_geometric.utils import degree
import networkx as nx
import pandas as pd
import matplotlib.cm as cm
from torch_geometric.data import Data
from torch_geometric.utils import from_networkx
import dataset
import re
import pickle
import random
```

```python
def make_correlated_graph(n_nodes=100):
    G = nx.gn_graph(n_nodes, seed=random.randint(0, 10**6))
    for node in G.nodes:
        if G.in_degree(node) == 2:
            G.nodes[node]["rule"] = "A and B" if node % 2 == 0 else "A or B"
    return G

graphs = [make_correlated_graph() for _ in range(300)]

with open("boolean_graphs_filtered.dat", "wb") as f:
    pickle.dump(graphs, f)

print(f"Saved {len(graphs)} Boolean graphs.")
```

```python
def parse_gate_label(rule_str):
    """Return 0 for AND, 1 for OR; -1 if mixed or unclear."""
    rule_str = rule_str.lower()
    has_and = "and" in rule_str
    has_or  = "or"  in rule_str
    if has_and and not has_or:
        return 0
    elif has_or and not has_and:
        return 1
    else:
        return -1  # mixed or unclear gate


def graph_to_training_data(G, feature_dim=8):
    """Convert one Boolean graph into x, edge_index, gate_triplets, gate_labels."""
    nodes = list(G.nodes)
    node_to_idx = {n: i for i, n in enumerate(nodes)}

    edges = list(G.edges)
    edge_index = torch.tensor(
        [[node_to_idx[u] for u, v in edges],
         [node_to_idx[v] for u, v in edges]],
        dtype=torch.long
    )

    # random features (you can later add structural features using your own data)
    x = torch.randn(len(nodes), feature_dim)

    gate_triplets, gate_labels = [], []
    for tgt in nodes:
        inputs = [src for src, _ in G.in_edges(tgt)]
        if len(inputs) == 2:
            rule = G.nodes[tgt].get("rule", "")
            label = parse_gate_label(rule)
            if label != -1:
                src1, src2 = inputs
                gate_triplets.append((node_to_idx[src1], node_to_idx[src2], node_to_idx[tgt]))
                gate_labels.append(label)

    return x, edge_index, gate_triplets, gate_labels


def load_boolean_dataset(dat_path="boolean_graphs_filtered.dat", feature_dim=8):
    
    """Load graphs and split into train/val/test (80/10/10). I did this to avoid scraping graphs, 
    this way you just select some entire graphs as training data and the other as test and validation"""
    
    graphs = pickle.load(open(dat_path, "rb"))
    random.shuffle(graphs)

    n_total = len(graphs)
    n_train = int(0.8 * n_total)
    n_val   = int(0.1 * n_total)

    splits = {
        "train": graphs[:n_train],
        "val":   graphs[n_train:n_train + n_val],
        "test":  graphs[n_train + n_val:]
    }

    data = {}

    for split_name, split_graphs in splits.items():
        all_x, all_edges, all_triplets, all_labels = [], [], [], []
        offset = 0

        for G in split_graphs:
            x, edge_index, local_triplets, local_labels = graph_to_training_data(G, feature_dim)
            if not local_triplets:
                continue

            n = x.size(0)
            shifted_triplets = [(a + offset, b + offset, c + offset) for (a, b, c) in local_triplets]
            shifted_edges = edge_index + offset

            all_x.append(x)
            all_edges.append(shifted_edges)
            all_triplets.extend(shifted_triplets)
            all_labels.extend(local_labels)
            offset += n

        if all_x:
            x_all = torch.cat(all_x, dim=0)
            edge_index_all = torch.cat(all_edges, dim=1)
        else:
            x_all = torch.empty(0, feature_dim)
            edge_index_all = torch.empty(2, 0, dtype=torch.long)

        data[split_name] = {
            "x": x_all,
            "edge_index": edge_index_all,
            "triplets": all_triplets,
            "labels": all_labels
        }

        print(f"{split_name.upper()}: {len(split_graphs)} graphs → nodes={x_all.size(0)}, gates={len(all_triplets)}")

    return data


# example use
data = load_boolean_dataset("boolean_graphs_filtered.dat", feature_dim=8)

x_train, edge_train = data["train"]["x"], data["train"]["edge_index"]
x_val, edge_val     = data["val"]["x"],   data["val"]["edge_index"]
x_test, edge_test   = data["test"]["x"],  data["test"]["edge_index"]

print("\nTrain triplets:", len(data["train"]["triplets"]))
print("Val triplets:", len(data["val"]["triplets"]))
print("Test triplets:", len(data["test"]["triplets"]))
```

```python
class GatePredictor(nn.Module):
    """
    Predicts the logic type (AND/OR) of a 2-input gate connecting two sources to a target.
    """
    def __init__(self, in_dim, hidden_dim=32):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(in_dim * 3, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1),
            nn.Sigmoid()
        )

    def forward(self, src1_emb, src2_emb, tgt_emb):
        z = torch.cat([src1_emb, src2_emb, tgt_emb], dim=-1)
        return self.mlp(z)

class GNNWithGateClassifier(nn.Module):
    def __init__(self, in_dim, hidden_dim, embed_dim, dropout=0.25):
        super().__init__()
        self.conv1 = GCNConv(in_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)
        self.gate_predictor = GatePredictor(embed_dim)

    def forward(self, x, edge_index, gate_triplets):

        x = F.relu(self.conv1(x, edge_index))
        x = self.dropout(x)
        x = F.relu(self.conv2(x, edge_index))

        preds = []
        for src1, src2, tgt in gate_triplets:
            p = self.gate_predictor(x[src1].unsqueeze(0),
                                    x[src2].unsqueeze(0),
                                    x[tgt].unsqueeze(0))
            preds.append(p)
        return torch.cat(preds, dim=0)
```

```python
# if you wanna use BH this line helps you speed up
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# prepare graphs
train_x = data["train"]["x"].to(device)
train_edge_index = data["train"]["edge_index"].to(device)
train_triplets = data["train"]["triplets"]
train_labels = torch.tensor(data["train"]["labels"], dtype=torch.float32).view(-1, 1).to(device)

val_x = data["val"]["x"].to(device)
val_edge_index = data["val"]["edge_index"].to(device)
val_triplets = data["val"]["triplets"]
val_labels = torch.tensor(data["val"]["labels"], dtype=torch.float32).view(-1, 1).to(device)

# ---- Model, optimizer, loss ----
model = GNNWithGateClassifier(in_dim=train_x.size(1), hidden_dim=32, embed_dim=16, dropout=0.5).to(device)
opt = torch.optim.Adam(model.parameters(), lr=6e-4, weight_decay=9e-4)
criterion = nn.BCEWithLogitsLoss()  # safer than BCELoss when using logits directly

# model setup
epochs = 200
train_losses, val_losses = [], []
best_val = float("inf")
patience, bad_epochs = 15, 0

print(f"Train gates: {len(train_triplets)}, Val gates: {len(val_triplets)}")

# Training
for epoch in range(epochs):
    # training
    model.train()
    opt.zero_grad()

    preds = model(train_x, train_edge_index, train_triplets)
    assert preds.shape == train_labels.shape, f"Shape mismatch: {preds.shape} vs {train_labels.shape}"

    loss = criterion(preds, train_labels)
    loss.backward()
    opt.step()
    train_losses.append(loss.item())

    # validation
    model.eval()
    with torch.no_grad():
        val_preds = model(val_x, val_edge_index, val_triplets)
        assert val_preds.shape == val_labels.shape, f"Shape mismatch: {val_preds.shape} vs {val_labels.shape}"

        val_loss = criterion(val_preds, val_labels)
        val_losses.append(val_loss.item())

        # compute accuracy (convert logits to probabilities)
        val_acc = ((torch.sigmoid(val_preds) > 0.5).float() == val_labels).float().mean().item()
        train_acc = ((torch.sigmoid(preds) > 0.5).float() == train_labels).float().mean().item()

    # record the logs
    if epoch % 10 == 0 or epoch == epochs - 1:
        print(f"Epoch {epoch:03d} | "
              f"Train Loss: {loss.item():.4f} | Val Loss: {val_loss.item():.4f} | "
              f"Train Acc: {train_acc:.3f} | Val Acc: {val_acc:.3f}")

    # early stopping to avoid overfitting
    if val_loss < best_val - 1e-4:
        best_val = val_loss
        bad_epochs = 0
        best_model_state = model.state_dict()
    else:
        bad_epochs += 1
        if bad_epochs >= patience:
            print(f"Early stopping at epoch {epoch}")
            break

# load the best model
model.load_state_dict(best_model_state)

# plot losses
plt.figure(figsize=(6,4))
plt.plot(train_losses, label='Train loss')
plt.plot(val_losses, label='Validation loss', linestyle='--')
plt.xlabel('Epoch')
plt.ylabel('BCE Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
